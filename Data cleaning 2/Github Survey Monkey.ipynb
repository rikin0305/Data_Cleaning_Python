# Calculating total respondents per question.

```python
d2['Questions'] = d2['Questions + Sub Questions'].map(lambda x : x.split("-")[0])
d3 = d2.copy()
d3
```

```python
d4 = d3[d3['Answers'].notna()]
d5 = d4.groupby("Questions")['Respondent ID'].nunique().reset_index()
d5 = d5.rename(columns={"Respondent ID":"Total Respondents"})
d5
```

```python
d6 = pd.merge(d3,d5,on="Questions")
d6
```

# Total no of people answering same answer per question

```python
d5 = d6.groupby(["Questions + Sub Questions","Answers"])['Respondent ID'].nunique().reset_index()
```

```python
d5
```

```python
d5 = d5.rename(columns={"Respondent ID":"Same Answer"})
d5
```

```python
d7 = pd.merge(d6,d5,on=["Questions + Sub Questions","Answers"],how="left")
d7['Same Answer'].fillna(0,inplace=True)
d7
```
