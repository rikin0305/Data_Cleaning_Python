```python
df = df.drop(["Unnamed: 9","Unnamed: 10"],axis="columns")
```

```python
df = df.dropna(how="all")
df
```

```python
df = df[df['Row Type'] != 'Row Type']
```

```python
df['iteration'] = l
```

```python
d = df[df['Row Type'].str.contains("first name")]
d = d.drop(['Speed1','Speed2','Electricity','Effort','Weight','Torque'], axis='columns')
d
```
